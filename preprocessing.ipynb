```python
import pydicom
import matplotlib.pyplot as plt

filePath = '/Users/pepo_abdo/Downloads/01_MRI_Data/0002/L-SPINE_CLINICAL_LIBRARIES_20160621_112938_873000/T1_TSE_TRA_0005/T1_TSE_TRA__0002_002.ima'
img = pydicom.read_file(filePath)

# Extract the pixel data from the DICOM object
pixel_array = img.pixel_array

# Create a figure and plot the image
fig = plt.figure()
plt.imshow(pixel_array, cmap=plt.cm.gray)
i = 0
# Set the save path for the image
save_path = 'LBP_XRAY/Image_{}.jpg'.format(i)

# Save the image
plt.imsave(save_path, pixel_array, cmap=plt.cm.gray)
plt.close(fig)

print(f"Image saved at: {save_path}")
```
